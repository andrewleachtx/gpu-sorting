```python
processes = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]

speedup_measurement = 'Max time/rank'
speedup_key = 1
def transform_df(df, arr):
    arr['speedup'] = {}
    for node in df.graph.traverse():
        node_str = str(node)
        name = node_str[node_str.index(':') + 3: node_str.index(',') - 1]

        if (name in arr):
            continue
        
        arr[name] = {}
        temp = arr[name]
        for size in sizes:

            temp[size] = {}

            if (name == 'main'):
                arr['speedup'][size] = {}
                
            for measurement in measurements:
                temp[size][measurement] = []

                if (name == 'main' and measurement == speedup_measurement):
                    arr['speedup'][size][measurement] = []

                for proc in processes:
                    try:
                        temp[size][measurement].append(df.dataframe.loc[node].loc[proc].loc[size][measurement])

                        if (name == 'main' and measurement == speedup_measurement):
                            arr['speedup'][size][measurement].append(df.dataframe.loc[node].loc[speedup_key].loc[size][measurement] / df.dataframe.loc[node].loc[proc].loc[size][measurement])
                    except:
                        temp[size][measurement].append(None)

                        if (name == 'main' and measurement == speedup_measurement):
                            arr['speedup'][size][measurement].append(None)
in_order_gage = {}            
random_gage = {}
reverse_gage = {}
pert_gage = {}

transform_df(sorted_gage_df, in_order_gage)
transform_df(random_gage_df, random_gage)
transform_df(reverse_gage_df, reverse_gage)
transform_df(pert_gage_df, pert_gage)

print('Andrew\n')

in_order_andrew = {}            
random_andrew = {}
reverse_andrew = {}


transform_df(sorted_andrew_df, in_order_andrew)
transform_df(random_andrew_df, random_andrew)
transform_df(reverse_andrew_df, reverse_andrew)
print('Anil\n')

in_order_anil = {}            
random_anil = {}
reverse_anil = {}

transform_df(sorted_anil_df, in_order_anil)
transform_df(random_anil_df, random_anil)
transform_df(reverse_anil_df, reverse_anil)

speedup_key = 2
processes = [3, 5, 9, 17, 33, 65, 129, 257, 513, 1024]
print('Brian\n')

in_order_brian = {}            
random_brian = {}
reverse_brian = {}

transform_df(sorted_brian_df, in_order_brian)
transform_df(random_brian_df, random_brian)
transform_df(reverse_brian_df, reverse_brian)

```

```python
def plot(arrays, names, title, yaxis='Time (s)'):
    fig, ax = plt.subplots(figsize=(15, 7))
    plt.xscale("log", base=2)
    plt.yscale('log', base=2)
    for name in names:
        ax.plot(processes, arrays[name], marker='o', label=name)
    ax.set_title(f'{title}')
    ax.set_xlabel('Number of Processes')
    ax.set_ylabel(yaxis)
    ax.legend()
    fig.patch.set_alpha(1.0)
    ax.patch.set_alpha(1.0)
    plt.savefig(title.replace(' ', '_'), bbox_inches='tight')
```
